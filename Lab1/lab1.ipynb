## Part 1: The Gravitational Potential of the Earth
----

The gravitational potential at a distance $r$ from a mass $M$ is given by:

$$\Phi(r) = -\frac{GM}{r}$$

where G is the gravitational constant.

1. Code up a function that takes as an input the mass of a body $M$ , the cartesian location of the
mass $(x_M , y_M)$, and the location of some distant evaluation point $(x, y)$. Have the function
return the potential at the evaluation point. You may consider approximating the Earth as
a point particle, but in which case, be careful! For in that approximation, the the potential
contains a singularity. In our function, use all modern best practices for coding.
E.g. type hints, type checks, pre and post conditions, doc strings, etc.

```python
def Potential(M,x_M,y_M,x,y):
    """ A function which calculates the gravitential potential at a distance from a mass of a body 

    args: 
        M (float): Mass of the body in kilograms
        x_M (float): Location of body in x axis of cartesian space in meters
        y_M (float): Location of body in y axis of cartesian space in meters
        x (float): Location of potential measurement in x axis of cartesian space in meters
        y (float): Location of potential measurement in y axis of cartesian space in meters

    Returns:
        float : Gravitational potential at distance from a body of mass
    """

    # Checking that the inputted variables make sense
    if not isinstance(M,float) and not isinstance(M,int):
        raise TypeError(f"Variable M must be float or integer.")
    if not M>0:
        raise ValueError(f"Variable M must be greater than 0.")
    if not isinstance(x_M,float) and not isinstance(x_M,int) and not isinstance(x_M,array):
        raise TypeError(f"Variable x_M must be float, integer, or array of numbers.")
    if not isinstance(y_M,float) and not isinstance(y_M,int) and not isinstance(y_M,array):
        raise TypeError(f"Variable y_M must be float, integer, or array of numbers.")

    # Calculate the potential
    r = np.sqrt((x-x_M)**2+(y-y_M)**2) 
    p = -G*M/r
    return np.where(r < 1,np.nan,p);
```

## Part 3: The Gravitational Force Field of the Earth-Moon System
----
The gravitational force $\vec{F}$ 
that a mass $M_1$ exerts on a mass $m_2$ is:

$$\vec{F}_{21} = -G \frac{M_1 m_2}{|\vec{r}_{21}|^2} \hat{r}_{21}$$

where $\vec{r}_{21}$ is the displacement vector from $M_1$ to $m_2$.

1. Code up a function that takes as an input the mass of a body $M_1$, 
the mass of a secondary body $m_2$,
the cartesian location of the first mass $(x_1, y_1)$, 
and the cartesian location of the second body $(x_2, y_2)$. 
Have the function return the two gravitational force components, 
$F_x$ and $F_y$, experienced by the mass $m_2$ 
from the force exerted by $M_1$.
Note that the gravitational force contains a singularity, 
and so if you made the approximation that the Earth is a finite point,
you will need to avoid it.
Like for the 1D potential, make sure it follows modern best-coding practices.

```python
def Gravity(M1,m2,x1,y1,x2,y2):
    """ A function which calculates the gravitational force one mass exerts on another

    args: 
        M1 (float): Mass of body 1 in kilograms
        m2 (float): Mass of body 2 in kilograms
        x1,y1 (float): Location of body 1 in cartesian space (x,y) in meters
        x2,y2 (float): Location of body 2 in cartesian space (x,y) in meters

    Returns:
        float : Gravitational force experienced by body 2 in x,y components
    """
    # Making sure input variables make sense
    if not isinstance(M1,float) and not isinstance(M,int):
        raise TypeError(f"Variable M1 must be float or integer.")
    if not M1>0:
        raise ValueError(f"Variable M1 must be greater than 0.")
    if not isinstance(x1,float) and not isinstance(x1,int) and not isinstance(x1,array):
        raise TypeError(f"Variable x1 must be float, integer, or array of numbers.")
    if not isinstance(y1,float) and not isinstance(y1,int) and not isinstance(y1,array):
        raise TypeError(f"Variable y1 must be float, integer, or array of numbers.")
    if not isinstance(x2,float) and not isinstance(x2,int) and not isinstance(x2,array):
        raise TypeError(f"Variable x2 must be float, integer, or array of numbers.")
    if not isinstance(y2,float) and not isinstance(y2,int) and not isinstance(y2,array):
        raise TypeError(f"Variable y2 must be float, integer, or array of numbers.")
        
    # Removing divide by zero issue
    if np.isclose(x1,x2) and np.isclose(y1,y2): 
        Fx = np.nan 
        Fy = np.nan 
    else:
        # Calculating force
        r_squared = np.hypot(x1-x2, y1-y2)**2
        vec_r = np.array([x1-x2, y1-y2])
        vec_r_hat = vec_r/np.linalg.norm(vec_r)
        Fx = G*(M1*m2) / r_squared * vec_r_hat[0]
        Fy = G*(M1*m2) / r_squared * vec_r_hat[1] 
    return Fx,Fy;
```

## Part 4: Altitude of the Saturn V Rocket
----
Rockets function by conservation of momentum.
The ejection of fuel backwards propels the rocket forwards.
The change in a rocket's velocity $\Delta v$ as a 
function of time $t$ (under a set of simplifying assumptions) 
is given by the Tsiolkovsky rocket equation:

$$\Delta v(t) = v_e \ln \left(\frac{m_0}{m(t)}\right) - gt$$

where $m_0$ is the initial "wet" mass (fuel + rocket parts + payload), 
$m(t) = m_0 - \dot{m}t$ is the mass at time $t$, 
$\dot{m}$ is the fuel burn rate (which we will assume to be constant), 
$v_e$ is the fuel exhaust velocity, and $g$ is the gravitational acceleration.

You can find the altitude $h$ of the rocket at "burnout" 
(when all the fuel is used up) by integrating the velocity over time:

$$h = \int_{0}^{T} \Delta v(t) \, dt$$

where $T$ is the total burn time of the rocket, 
which by conservation of momentum arguments is:

$$T = \frac{m_0 - m_f}{\dot{m}}$$

where $m_f$ is the final "dry mass" of the rocket, 
once all the fuel is burned.

1. Calculate the burn time $T$ for the first stage of
the new Saturn V rocket that NASA is planning
to use to carry the Apollo capsule.

```python
# Change in mass
change = m_0 - m_f 

# Burn Time Equation
T = change/m_dot

print(T,"seconds")
```

2. Define a function that returns the change in velocity $\Delta v(t)$.
It should take all necessary arguments, meaning: the time, 
wet mass, dry mass, burn rate, exhaust speed, and gravitational constant $g$. 
Have it return the change in velocity.
Note that the rocket eventually runs out of fuel, 
so you should design your function carefully so that 
$\Delta v$ eventually becomes zero after all fuel is spent.
Again, document and guard your function according to best practices.

```python
def velocity(t,m_0,m_f,m_dot,v_e):
    """ A function which the velocity of a rocket at any given time during its launch

    args: 
        t (float): Time since rocket launch start in seconds
        m_0 (float): Wet mass of the rocket (structure + fuel) in kilograms
        m_f (float): Dry mass of the rocket (structure) in kilograms
        m_dot (float): Fuel burn rate in kilograms per second
        v_e (float): Fuel exhaust velocity in meters per second

    Returns:
        float : Velocity of the rocket at the given time in meters per second
    """
    # Checking variables make sense
    # t
    if not isinstance(t,float) and not isinstance(t,int) and not isinstance(t,array):
        raise TypeError(f"Variable t must be float, integer, or array of numbers.")
    # m_0
    if not isinstance(m_0,float) and not isinstance(m_0,int):
        raise TypeError(f"Variable m_0 must be float or integer.")
    if not m_0>0:
        raise ValueError(f"Variable m_0 must be greater than 0.")
    # m_f
    if not isinstance(m_f,float) and not isinstance(m_f,int):
        raise TypeError(f"Variable m_f must be float or integer.")
    if not m_f>0:
        raise ValueError(f"Variable m_f must be greater than 0.")
    # m_dot
    if not isinstance(m_dot,float) and not isinstance(m_dot,int):
        raise TypeError(f"Variable m_dot must be float or integer.")
    if not m_dot>0:
        raise ValueError(f"Variable m_dot must be greater than 0.")
    # v_e
    if not isinstance(v_e,float) and not isinstance(v_e,int):
        raise TypeError(f"Variable v_e must be float or integer.")
    if not v_e>0:
        raise ValueError(f"Variable v_e must be greater than 0.")

    # Calculating velocity
    m_t = m_0 - m_dot*t 
    log = np.log(m_0/m_t)
    v = v_e*log-G*t 
    return v
```
